## MM data Read and Preprocessing

```python
## The MM dataset
txt_path = "D:\\workspace\\DeepLearning_codes\\AlphaReal\\MM_raw.txt"
df = pd.read_csv(txt_path, sep='\t')
print(df.head())
```

## pandas.core.window.Apply

```python
M_window_size = 3 # month

def div_func(array):
    return ((array[-1]/array[0])-1)*100

df_MM = df.rolling(M_window_size, center=True).apply(lambda x: div_func(x)).dropna()
```

```python
df_MM
```

## merge dataframe

```python
df_MM['Date Time'] = df_MM.index
df_permit['Date Time'] = df_permit.index
```

```python
df_merged = pd.merge(df_MM, df_permit, on='Date Time', suffixes=('_M','_P'))
```

```python
df_merged.index = df_merged['Date Time']
```

```python
df_merged.columns
```

```python
df_ = pd.DataFrame(index=['0', '1'], columns=['MM', 'Permits'])
df_ = df_.fillna(np.NaN)
df_
```

```python
for reg in list(df.columns):
    reg_M = reg+"_M"
    reg_P = reg+"_P"   
    merged_reg = pd.merge(df_merged[reg_M], df_merged[reg_P], left_index=True, right_index=True)    
    df_ = pd.concat([df_,merged_reg.rename(columns={reg_M:'MM', reg_P:'Permits'})])
df_merged_stack = df_.dropna()
```

# 2. Correlation

## Pearson correlation

```python
import seaborn as sns
import scipy.stats as stats
%matplotlib inline
mpl.rcParams['figure.figsize'] = (8, 6)
mpl.rcParams['axes.grid'] = False
```

```python
df = df_merged_stack[['Permits','MM']]
```

```python
overall_pearson_r = df.corr().iloc[0,1]
print(f"Pandas computed Pearson r: {overall_pearson_r}")
```

```python
r, p = stats.pearsonr(df.dropna()['Permits'], df.dropna()['MM'])
print(f"Scipy computed Pearson r: {r} and p-value: {p}")
```
